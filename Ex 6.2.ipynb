## Q3 Create a correlation matrix heatmap (colored).

```python
# Create a correlation heatmap using matplotlib

plt.matshow(numeric_df.corr())
plt.show()
```

```python
# Save figure
plt.matshow(numeric_df.corr())
plt.savefig(os.path.join(path, '04 Visualization', 'correlation_heatmap.png'))
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(numeric_df.corr(), fignum=f.number) # type of plot
plt.xticks(range(numeric_df.shape[1]), numeric_df.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(numeric_df.shape[1]), numeric_df.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(numeric_df.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

```python
#Save fig
f.savefig(os.path.join(path, '04 Visualization', 'correlation_heatmap_with_title.png'))
```

Interestingly, the given dataset shows a moderate relationship between rent and other factprs like bathroom, size and BHK. 
The strongest corelation is for Size and BHK, Bathroom and BHK.

## Q6 Create a categorical plot and interpret the results

```python
# Create a categorical plot in seaborn using the categorical variable "Furnishing Status" and numerical variable "Rent" and "Size"

sns.set(style="ticks")
l = sns.catplot(x="BHK", y="Rent", hue="Furnishing Status", data=df_rent)
```

```python
#Save fig
l.savefig(os.path.join(path, '04 Visualization', 'catplot_furnishing.png'))
```

You can see that listings between 3 and 4 BHK have higher "Semi Furnished" furnishings. Also, the rents tend to be higher for Semi furnished and Furnished furnishing status.


```python
# Create a categorical plot in seaborn using the categorical variable "City" and numerical variable "Rent" and "BHK"

sns.set(style="ticks")
k = sns.catplot(x="BHK", y="Rent", hue="City", data=df_rent)
```

Notice that "Mumbai" has the highest rents in every category of the apartment. "Hyderabad" tends to have rents more on the lower side.

```python
#Save fig
k.savefig(os.path.join(path, '04 Visualization', 'catplot_city.png'))
```
